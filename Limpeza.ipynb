```python
import pandas as pd

df = pd.read_csv('tmdb_movies_detailed.csv', delimiter=',')
df = df.drop('sinopse', axis=1)
df.head()
print(df.isnull().sum())
```

```python
linhas_com_nulos = df[df.isnull().any(axis=1)]

print("Linhas com pelo menos um valor nulo:")
print(linhas_com_nulos)
```

```python
df = df.dropna()
print(df.isnull().sum())
num_ids_unicos = df['id_filme'].nunique()

print(f"O número de IDs de filmes únicos é: {num_ids_unicos}")
```

```python
output_filename_limpo = 'filme_limpo.csv'
df.to_csv(output_filename_limpo, index=False, encoding='utf-8')
```
